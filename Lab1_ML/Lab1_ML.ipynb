# Activity 2 : Read, Write File

Đọc tệp

```python
with open("test.txt","r") as f:
    for line in f:
        print(line)
```

Ghi vào tệp (‘w’):

```python
with open("b.txt","w") as f:
    f.write("Hello World !!!")
    
with open("b.txt","r") as f:
    contents = f.read()
    
print(contents)
```

Ghi thêm vào cuối tệp (‘a’):

```python
with open("b.txt","a") as f:
    f.write("\nHello Again")
    
with open("b.txt","r") as f:
    contents = f.read()
    
print(contents)
```

Xóa tệp văn bản

```python
import os
os.remove("b.txt")
```

Các chế độ : w+, r+, a+

```python
# Mở tệp văn bản để ghi và đọc
with open("b.txt","w+") as f:
    # Ghi nội dung vào tệp
    f.write("Hello World !!!")
    
    # Đóng tệp
    f.close()
```

```python
with open("b.txt","r+") as f:
    # Thêm nội dung vào cuối tệp
    f.write("Hello Again !!!")
    
    # Đóng tệp
    f.close()
```

```python
# Mở tệp văn bản để đọc và ghi
with open("test.txt","a+") as f:
    # Đọc nội dung tệp
    contents = f.read()
    
    # Thêm nội dung vào đầu tệp
    f.write("\n")
    f.write("New Content")
    
    # Đóng tệp
    f.close()
```
